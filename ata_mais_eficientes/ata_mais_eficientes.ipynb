## Importando os dados

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_tables = pd.read_excel("xG.xlsx")
```

```python
df_tables
```

## Removendo colunas desnecessárias

```python
df_tables = df_tables.drop(columns=['Nação', 'Idade', 'Nascimento', 'Class.', 'FK', 'PB', 'PT', 'Partidas', ])
```

```python
df_tables.head()
```

```python
df_tables = df_tables.drop(0)
```

```python
df_tables.head()
```

```python
print(df_tables.columns)
```

## Filtrando os atacantes

```python
df_ata = df_tables[df_tables['Pos.'] == 'AT']
```

```python
df_ata = df_ata.loc[:,['Jogador','Equipe','90s','G-xG']]
```

```python
df_ata = df_ata.sort_values('G-xG',  ascending=False)
```

```python
df_ata
```

## Atacantes mais eficientes

```python
df_positivo = df_ata[df_ata['G-xG'] > 0 ]
```

```python
plt.figure(figsize=(25, 5))
plt.bar(df_positivo['Jogador'], df_positivo['G-xG'], color = 'green')
plt.xticks(df_positivo['Jogador'])
plt.xlabel('Jogador')
plt.ylabel('G-xG')
plt.title('Atacantes mais eficientes')
plt.show()
```

## Atacantes menos eficientes

```python
df_negativo = df_ata[df_ata['G-xG'] <= 0 ]
```

```python
df_negativo = df_negativo.sort_values('G-xG',  ascending=True)
```

```python
plt.figure(figsize=(40, 10))
plt.bar(df_negativo['Jogador'], df_negativo['G-xG'], color = 'red')
plt.xticks(df_negativo['Jogador'])
plt.xlabel('Jogador')
plt.ylabel('G-xG')
plt.title('Atacantes menos eficientes')
plt.show()
```
