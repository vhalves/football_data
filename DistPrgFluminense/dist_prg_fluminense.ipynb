## Removendo colunas desnecessárias e renomeando as mais importantes

```python
df = df.drop(columns=['Unnamed: 1_level_0','Unnamed: 0','Unnamed: 27_level_0','Unnamed: 3_level_0'])
```

```python
df_posse = df.rename(columns={'Unnamed: 0_level_0':'Jogador','Unnamed: 4_level_0':'90s','Conduções.2':'DistPrg'})
```

```python
df_final = df_posse.drop([0, 38, 39])
```

```python
df_prog = df_final.loc[:,['Jogador','90s','DistPrg']]
```

```python
df_prog
```

## Calculando a distância que cada jogador progride por 90 minutos

```python
df_prog['90s'] = df_prog['90s'].astype('float')
df_prog['DistPrg'] = df_prog['DistPrg'].astype('float')
```

```python
df_prog['DistPrg/90s'] = df_prog['DistPrg']/df_prog['90s']
```

```python
df_prog
```

```python
df_prog = df_prog.sort_values('DistPrg/90s',  ascending=False)
```

```python
## Ranking de jogadores (com minutagem relevante) que mais progridem com a bola

df_prog[df_prog['90s'] >= 8]
```
